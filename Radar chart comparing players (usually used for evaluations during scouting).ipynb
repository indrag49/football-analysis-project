```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from soccerplots.radar_chart import Radar
```

```python
stats = pd.read_csv("2020-21 La Liga player stats (per 90s) - Sheet1.csv")
stats
```

```python
stats['Player'] = stats['Player'].str.split('\\')
stats
```

```python
names = stats['Player']
names = pd.DataFrame(names.to_list(), columns=['name1', 'name2'])
names
```

```python
stats['Player'] = names['name1']
stats
```

```python
stats.Player.unique()
```

```python
S = stats[(stats['Player'] == 'Luka Modrić') | (stats['Player'] == 'Toni Kroos')]
S
```

```python
S.columns
```

```python
S = S[['Player', 'Gls', 'Ast', 'G+A', 'G-PK', 'G+A-PK', 'xG', 'xA',
       'xG+xA', 'npxG', 'npxG+xA']].reset_index()
S
```

```python
S = S.rename(columns={"Gls": "Goals", "Ast": "Assists", "G+A": "Goals+Assists", "G-PK": "Non Penalty Goals", "G+A-PK": "Non Penalty Goals+Assists",
                     "xG": "Expected Goals", "xA": "Expected Assists", "npxG": "Non Penalty Expected Goals", "npxG+xA": "Non Penalty Expected Goals+Expected Assists"})
S
```

```python
S.columns
```

```python
S = S[['Player', 'Goals', 'Assists', 'Goals+Assists',
       'Non Penalty Goals', 'Non Penalty Goals+Assists', 'Expected Goals',
       'Expected Assists', 'xG+xA', 'Non Penalty Expected Goals',
       'Non Penalty Expected Goals+Expected Assists']]
S
```

```python
col_names = S.columns.to_list()
col_names
```

```python
c = col_names[1:]
c
```

```python
R = []
for colnames in col_names[1:]:
    Min = min(S[colnames])
    Max = max(S[colnames])
    R.append((Min - 0.2 * Min, Max + 0.2 * Max))
R
```

```python
S1 = []
for i in range(len(S)):
    S1 += [S.T[i].tolist()[1:], ]
S1
```

```python
T = dict(
    title_name = 'Toni Kroos',
    title_color = 'blue',
    subtitle_name = 'MF, Real Madrid',
    subtitle_color = 'blue',
    title_name_2 = 'Luka Modrić',
    title_color_2 = 'red',
    subtitle_name_2 = 'MF, Real Madrid',
    subtitle_color_2 = 'red'
)
T
```

```python
note = '@indraghosh314 \nFBREF data'
```

```python
radar = Radar(background_color="black", patch_color="#28252C", label_color="white", range_color="#BFE9BF")
fig, ax = radar.plot_radar(ranges = R, params = c, values = S1, alphas = [0.76, 0.6], 
                           title = T, endnote = note, radar_color=['#0f4c75', '#e94560'], compare=True)
```
