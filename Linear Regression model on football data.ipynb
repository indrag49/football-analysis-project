```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

### Get the csv for shots data

```python
shots = pd.read_csv("La Liga 2020-21 shot stats - Sheet1.csv")
```

```python
shots
```

```python
shots['Player'] = shots['Player'].str.split('\\')
shots
```

```python
names = shots['Player']
names = pd.DataFrame(names.to_list(), columns=['name1', 'name2'])
names
```

```python
shots['Player'] = names['name1']
shots
```

```python
shots.columns
```

```python
shots = shots[shots['Pos']!='GK'].reset_index()
shots
```

```python
sns.pairplot(shots[['90s', 'Gls', 'Sh', 'SoT', 'Dist']])
```

### As expected *Gls* (Gaols) has high correlation with *Sh* (shots) and *SoT* (Shots on Target). We can check which one has the higher correlation

```python
abs(shots['Gls'].corr(shots['Sh']))
```

```python
abs(shots['Gls'].corr(shots['SoT']))
```

### Obviously the correlation between *Gls* and *SoT* is higher

```python
x = shots['SoT'].values.reshape((-1, 1))
y = shots['Gls'].values
```

```python
x.shape, y.shape
```

### Now we create the linear regression model and fit the data

```python
model = LinearRegression().fit(x, y)
```

### We can get the coefficient of determination $r^2 \in [0, 1]$. This tell us how good our model is. A value of $1$ tells that there is a perfect fit

```python
r2 = model.score(x, y)
r2
```

### Next we will determine the model parameters $m$ and $c$ for drawing the regression line of our model, such that the equation of the line is given by $y=mx+c$, where $m$ is the slope of the line and $c$ is the intercept

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

### Now we can perform predictions

```python
pred = model.predict(x)
pred
```

```python
fig, ax = plt.subplots(figsize=(13.5, 8))
#fig.set_facecolor(('black'))
ax.patch.set_facecolor(('black'))

plt.scatter(x, y, color='red')
plt.plot(x, pred, color='blue', linewidth=5)

plt.xlabel('Shots on Target by a player', size=20)
plt.ylabel('Goals by a player', size=20)
plt.title("La Liga Goals Vs. Shots on Target", size=20)
```

```python
sns.displot((y-pred),bins=50, color = 'red')
```

### Let us create a random dataset of player shots and perform some predictions on them

```python
test_x = np.random.randint(1, 90, 20)
test_x
```

```python
def predict_y(x):
    return m[0]*x + c

test_y = predict_y(test_x)
test_y
```

```python
pred_df = pd.DataFrame({'test_x':test_x, 'test_y':test_y})
pred_df
```
