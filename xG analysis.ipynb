```python
events = sb.events(match_id = 7576)
events
```

```python
events.columns
```

```python
E1 = events[['location', 'minute', 'player', 'team', 'shot_outcome', 'shot_statsbomb_xg', 'shot_technique', 'shot_type']]
E1
```

```python
E1 = E1[E1['shot_outcome'].isnull()==False].reset_index()
E1
```

```python
shots_Loc = E1['location']
shots_Loc = pd.DataFrame(shots_Loc.to_list(), columns=['x', 'y'])
shots_Loc
```

```python
E1['x'] = shots_Loc['x']
E1['y'] = shots_Loc['y']
E1
```

```python
E1 = E1[['minute', 'player', 'team', 'shot_outcome', 'shot_statsbomb_xg', 'shot_technique', 'shot_type', 'x', 'y']]
E1
```

```python
E1_Spain = E1[E1['team'] == 'Spain'].reset_index()
E1_Portugal = E1[E1['team'] == 'Portugal'].reset_index()
E1_Spain
```

```python
E1_Portugal
```

```python
# Extract the expected goals and the time data for both the teams
Spain_xg = E1_Spain['shot_statsbomb_xg'].tolist()
Spain_minute = E1_Spain['minute'].tolist()

Portugal_xg = E1_Portugal['shot_statsbomb_xg'].tolist()
Portugal_minute = E1_Portugal['minute'].tolist()
```

```python
# generate the cumulative expected goals values
Spain_xg_cumu = np.cumsum(Spain_xg)
Portugal_xg_cumu = np.cumsum(Portugal_xg)
```

```python
Spain_minute = [0] + Spain_minute
Portugal_minute = [0] + Portugal_minute
Spain_xg_cumu = [0] + Spain_xg_cumu.tolist()
Portugal_xg_cumu = [0] + Portugal_xg_cumu.tolist()
```

```python
Spain_minute, Spain_xg_cumu
```

```python
# plot the time chart for the cumulative expected goals
fig, ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor(('#8A8A8A'))
ax.patch.set_facecolor(('#8A8A8A'))

plt.xticks(range(0, 105, 15))
plt.xlabel('Time in Minutes')
plt.ylabel('xG')

ax.step(Spain_minute, Spain_xg_cumu, where='post', color='red', label='Spain xG', linewidth=6)
ax.step(Portugal_minute, Portugal_xg_cumu, where='post', color='white', label='Portugal xG', linewidth=6)
ax.legend(borderpad=1, markerscale=0.5, labelspacing=1.5, loc='upper center', fontsize=15)

ax.set_title("Timing Chart of xG values for Spain vs Portugal (3-3) match [World Cup 2018]", size=25)
```

```python
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='grass', line_color='#c7d5cc', figsize=(13.5, 8),
              constrained_layout=True, tight_layout=False)
fig, ax = pitch.draw()
#plt.gca().invert_yaxis()

for i in range(len(E1_Spain)):
    if E1_Spain.shot_outcome[i]=='Goal':
        pitch.scatter(E1_Spain.x[i], E1_Spain.y[i], marker='football', s=1000*E1_Spain.shot_statsbomb_xg[i], ax=ax)
    elif E1_Spain.shot_outcome[i]=='Blocked':
        pitch.scatter(E1_Spain.x[i], E1_Spain.y[i], color='red', s=1000*E1_Spain.shot_statsbomb_xg[i], ax=ax)
    elif E1_Spain.shot_outcome[i] in ['Wayward', 'Off T']:
        pitch.scatter(E1_Spain.x[i], E1_Spain.y[i], color='orange', s=1000*E1_Spain.shot_statsbomb_xg[i], ax=ax)
    else:
        pitch.scatter(E1_Spain.x[i], E1_Spain.y[i], color = "magenta", s=1000*E1_Spain.shot_statsbomb_xg[i], ax=ax)
ax.set_title("Spain Shots, xG:"+str(Spain_xg_cumu[-1])+", av.xG:"+str(Spain_xg_cumu[-1]/len(E1_Spain)), size=20)
```
